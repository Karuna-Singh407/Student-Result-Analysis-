```python
def classify_grade(cpi):
    if cpi >= 8.5:
        return 'A (Excellent)'
    elif cpi >= 7.5:
        return 'B (Good)'
    elif cpi >= 6.5:
        return 'C (Average)'
    else:
        return 'D (Need Improvement)'

df['Grade'] = df['CPI'].apply(classify_grade)
```

```python
def classify_grade(spi):
    if spi >= 8.5:
        return 'A (Excellent)'
    elif spi >= 7.5:
        return 'B (Good)'
    elif spi >= 6.5:
        return 'C (Average)'
    else:
        return 'D (Need Improvement)'

df['Grade1'] = df['SPI S1'].apply(classify_grade)
df['Grade2'] = df['SPI S2'].apply(classify_grade)
df['Grade3'] = df['SPI S3'].apply(classify_grade)
df['Grade4'] = df['SPI S4'].apply(classify_grade)
df['Grade5'] = df['SPI S5'].apply(classify_grade)
```

```python
#CPI is classified into grades
```

```python
#Data is visualised based on grade classification 
```

```python
plt.figure(figsize =(15,20))
plt.subplot(3,2,1)
sns.countplot(data=df, x='Grade', order=['A (Excellent)', 'B (Good)', 'C (Average)', 'D (Poor)'], palette='RdYlBu_r')
plt.title('Student Grade Distribution Based on CPI')
plt.xlabel('Grade')
plt.ylabel('Number of Students')

plt.subplot(3,2,2)
sns.countplot(data=df, x='Grade1', order=['A (Excellent)', 'B (Good)', 'C (Average)', 'D (Poor)'], palette='RdYlBu_r')
plt.title('Student Grade Distribution Based on SPI S1')
plt.xlabel('Grade')
plt.ylabel('Number of Students')

plt.subplot(3,2,3)
sns.countplot(data=df, x='Grade2', order=['A (Excellent)', 'B (Good)', 'C (Average)', 'D (Poor)'], palette='RdYlBu_r')
plt.title('Student Grade Distribution Based on SPI S2')
plt.xlabel('Grade')
plt.ylabel('Number of Students')

plt.subplot(3,2,4)
sns.countplot(data=df, x='Grade3', order=['A (Excellent)', 'B (Good)', 'C (Average)', 'D (Poor)'], palette='RdYlBu_r')
plt.title('Student Grade Distribution Based on SPI S3')
plt.xlabel('Grade')
plt.ylabel('Number of Students')

plt.subplot(3,2,5)
sns.countplot(data=df, x='Grade4', order=['A (Excellent)', 'B (Good)', 'C (Average)', 'D (Poor)'], palette='RdYlBu_r')
plt.title('Student Grade Distribution Based on SPI S4')
plt.xlabel('Grade')
plt.ylabel('Number of Students')

plt.subplot(3,2,6)
sns.countplot(data=df, x='Grade5', order=['A (Excellent)', 'B (Good)', 'C (Average)', 'D (Poor)'], palette='RdYlBu_r')
plt.title('Student Grade Distribution Based on SPI S5')
plt.xlabel('Grade')
plt.ylabel('Number of Students')
plt.savefig("Grade Distribution.jpg", dpi=2000, transparent=True) 
plt.show()
```
